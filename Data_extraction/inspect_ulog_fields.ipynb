```python
from pyulog import ULog
```

```python
def print_all_fields(ulog):
    """
    Print all fields for all datasets (dataframes) in a ULog file.

    :param ulog: The loaded ULog object
    """
    # Get all datasets
    data = ulog.data_list

    # Iterate through all datasets and print their fields
    for dataset in data:
        print(f"Dataset Name: {dataset.name}")  # Print the dataset name
        print("-" * 50)  # Separator
        # Print all fields in the dataset
        for field in dataset.field_data:
            print(f"{field.type_str:10} {field.field_name}")
        print("=" * 80)  # Separator for the next dataset
        print()
        print()
```

```python
def main():
    """
    Main function to load the ULog file and print all fields.
    """
    # Path to the ULog file
    file_name = '/home/hawk/GitDir/Logging/2024-12-09/12_40_59.ulg'  # Replace with the path to your ULog file

    # Load the ULog file
    try:
        ulog = ULog(file_name)
        print(f"Successfully loaded ULog file: {file_name}")
    except Exception as e:
        print(f"Error loading ULog file: {e}")
        return

    # Print all fields in the ULog file
    print_all_fields(ulog)
```

```python
if __name__ == "__main__":
    main()
```
